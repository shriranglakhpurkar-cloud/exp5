```python
car={
    "brand":"ford","model":"mustang","year":1964,
}
car.update({"color":"red"})
print(car)
x=car.values()
print(x)
```

```python
marks={
    "amit":"10","rahul":"5","rohit":"20",
}
name=input("enter student name")
print(marks.get(name,"student not found"))
```

```python
marks={
    "amit":"50","rahul":"60","rohit":"20",
}
topper=max(marks,key=marks.get)
print("topper:",topper)
print("marks:",marks[topper])
```
